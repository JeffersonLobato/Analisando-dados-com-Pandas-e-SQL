### Challenge: Create a table containing a time series with the absolute and percentage difference between the cheapest and most expensive values. Also present the states in which the highest and lowest prices were recorded next to it.

### DESAFIO: Crie uma tabela contendo uma serie temporal com a diferença absoluta e percentual entre os valores mais baratos e caros. Apresente também ao lado os estados na qual os maiores e menores preços foram registrados.

```python
df_max = gas_2000_2010.groupby('ANO E MÊS').max()['PREÇO MÉDIO REVENDA']
df_min = gas_2000_2010.groupby('ANO E MÊS').max()['PREÇO MÍNIMO REVENDA']
```

```python
idx_max = gas_2000_2010.groupby('ANO E MÊS')['PREÇO MÉDIO REVENDA'].idxmax()
idx_min = gas_2000_2010.groupby('ANO E MÊS')['PREÇO MÉDIO REVENDA'].idxmin()
```

```python
df_diff = pd.DataFrame()
```

```python
df_diff['DIFERENÇA ABSOLUTA'] = df_max-df_min
df_diff['DIFERENÇA PERCENTUAL'] = (df_max-df_min)/df_min*100
df_diff['PREÇO MÁXIMO'] = df_max
df_diff['PREÇO MÍNIMO'] = df_min
```

```python
df_diff['ESTADO MAIOR PREÇO'] = gas_2000_2010.loc[idx_max, :]['ESTADO'].values
df_diff['ESTADO MENOR PREÇO'] = gas_2000_2010.loc[idx_min, :]['ESTADO'].values
```
