```python
import pandas as pd
from numpy.random import randn

df = pd.DataFrame(randn(5,4), index=["A", "B" ,"C" ,"D" , "E"], columns="W X Y Z".split())
df
```

```python
#fetch column names

df.columns
```

```python
#search index names

df.index
```

```python
#Create new indexes, transforming previous indexes into a column

df.reset_index()
```

```python
df
```

```python
#To make the change permanent, you must use inplace=True

df.reset_index(inplace=True)
df
```

```python
#Converting a column to an index
#in this case the new index column will return to being the index

df.set_index('index', inplace=True)
df
```

```python
#Creating a column to be my new index

new_index = 'BR PR DF RR AM'.split()
new_index
```

```python
df['NEW INDEX'] = new_index
df
```

```python
#creating the new index from an existing column, however note that the previous index is deleted

df.set_index('NEW INDEX')
```

```python
#For the previous index to become a new column when 
#transforming another column into an index, I can use reset_index()

df.reset_index().set_index('NEW INDEX')
```
