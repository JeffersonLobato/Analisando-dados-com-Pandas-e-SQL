<h1>Merge, concat e join</h1>

<p>Methods that allow to merge DataFrames.</p>

<p>MÃ©todos que permitem combinar DataFrames.</p>

```python
import pandas as pd

df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                    'B': ['B0', 'B1', 'B2', 'B3'],
                    'C': ['C0', 'C1', 'C2', 'C3'],
                    'D': ['D0', 'D1', 'D2', 'D3']},
                      index= [0, 1, 2, 3])


df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
                    'B': ['B4', 'B5', 'B6', 'B7'],
                    'C': ['C4', 'C5', 'C6', 'C7'],
                    'D': ['D4', 'D5', 'D6', 'D7']},
                      index= [4, 5, 6, 7])


df3 = pd.DataFrame({'A': ['A8', 'A9', 'A10', 'A11'],
                    'B': ['B8', 'B9', 'B10', 'B11'],
                    'C': ['C8', 'C9', 'C10', 'C11'],
                    'D': ['D8', 'D9', 'D10', 'D11']},
                      index= [8, 9, 10, 11])
```

<h2>Concat</h2>

```python
df1
```

```python
df2
```

```python
df3
```

```python
#I can concatenate DataFrames, the method below by default stacks one DataFrame on top of the other.

pd.concat([df1, df2, df3])
```

```python
#I can concatenate DataFrames by inserting them side by side instead of below using the axis parameter.
#The following case is not ideal since our DataFrames share columns and not indexes, 
#but let's see how it looks.


pd.concat([df1, df2, df3], axis=1)
```

<h2>Merge</h2>

<p>The merge method in Pandas is used to merge two or more DataFrames based on one or more common columns.</p>

```python
left = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                     'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3']})
   
right = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                      'C': ['C0', 'C1', 'C2', 'C3'],
                      'D': ['D0', 'D1', 'D2', 'D3']})
```

```python
left
```

```python
right
```

```python
pd.merge(left, right, on='key')
```

```python
#Now in these DataFrames, the columns key1 and key2 exist in both, but they do not have the same data.


left = pd.DataFrame({'key1': ['K0', 'K0', 'K1', 'K2'],
                     'key2': ['K0', 'K1', 'K0', 'K1'],
                     'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3']})
    
right = pd.DataFrame({'key1': ['K0', 'K1', 'K1', 'K2'],
                      'key2': ['K0', 'K0', 'K0', 'K0'],
                      'C': ['C0', 'C1', 'C2', 'C3'],
                      'D': ['D0', 'D1', 'D2', 'D3']})
```

```python
left
```

```python
right
```

```python
#Now it will return only the rows that have matching occurrences in both DataFrames and merge the columns.
#In this example, it will take K0 K0, K1 K0 (twice, so two lines).

pd.merge(left, right, on=['key1', 'key2'])
```

```python
#If I use the how parameter with the outer value, it will bring all rows from both DataFrames,
#and where there are no corresponding elements in both DataFrames, it will fill with NaN.

pd.merge(left, right, how='outer', on=['key1','key2'])
```

```python
#If I use the parameter "how" with the value "left", 
#it will bring all the rows from the left DataFrame and 
#if there is no corresponding element in the right DataFrame, 
#in this case columns C and D, it will fill with NaN.

pd.merge(left, right, how='left', on=['key1','key2'])
```

```python
#If I use the "how" parameter with the "right" value, 
#it will bring all the rows from the right DataFrame, and 
#if there are no corresponding elements in the left DataFrame, 
#in this case columns A and B, it fills with NaN.

pd.merge(left, right, how='right', on=['key1','key2'])
```

<h2>Join</h2>

<p>The join method in Pandas is used to combine two DataFrames into a single one based on their common index or columns. It can be used to perform inner, outer, left, and right joins.</p>

```python
left = pd.DataFrame({'A': ['A0', 'A1', 'A2'],
                     'B': ['B0', 'B1', 'B2']},
                      index=['K0', 'K1', 'K2']) 

right = pd.DataFrame({'C': ['C0', 'C2', 'C3'],
                    'D': ['D0', 'D2', 'D3']},
                      index=['K0', 'K2', 'K3'])
```

```python
left
```

```python
right
```

```python
#Join is similar to merge, but it is applied directly to the DataFrame,
#for example, to join the right DataFrame to the left DataFrame.


left.join(right)
```

```python
#I can also use the parameter "how".

left.join(right, how='outer')
```
