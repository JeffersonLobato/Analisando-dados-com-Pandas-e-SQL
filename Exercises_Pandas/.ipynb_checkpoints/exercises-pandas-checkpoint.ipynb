### DESAFIO: Crie uma tabela contendo uma serie temporal com a diferença absoluta e percentual entre os valores mais baratos e caros. Apresente também ao lado os estados na qual os maiores e menores preços foram registrados.

```python
gas_2000_2010['PREÇO MÁXIMO DISTRIBUIÇÃO'] = pd.to_numeric(gas_2000_2010['PREÇO MÁXIMO DISTRIBUIÇÃO'], errors='coerce')
gas_2000_2010['PREÇO MÍNIMO DISTRIBUIÇÃO'] = pd.to_numeric(gas_2000_2010['PREÇO MÍNIMO DISTRIBUIÇÃO'], errors='coerce')
```

```python
gas_2000_2010['PREÇO MÁXIMO DISTRIBUIÇÃO'].dtypes
```

```python
gas_2000_2010['PREÇO MÍNIMO DISTRIBUIÇÃO'].dtypes
```

```python
price_dif_rev = gas_2000_2010['PREÇO MÁXIMO REVENDA']-gas_2000_2010['PREÇO MÍNIMO REVENDA']
price_dif_dist = gas_2000_2010['PREÇO MÁXIMO DISTRIBUIÇÃO']-gas_2000_2010['PREÇO MÍNIMO DISTRIBUIÇÃO']
```

```python
perc_rev = (price_dif_rev*100)/gas_2000_2010['PREÇO MÍNIMO REVENDA']
perc_dist = (price_dif_dist*100)/gas_2000_2010['PREÇO MÍNIMO DISTRIBUIÇÃO']
```

```python
new_df_gas = gas_2000_2010.loc[:,['DATA INICIAL','ESTADO','PREÇO MÍNIMO REVENDA','PREÇO MÁXIMO REVENDA','PREÇO MÍNIMO DISTRIBUIÇÃO','PREÇO MÁXIMO DISTRIBUIÇÃO']]
```

```python
new_df_gas['DIFEREÇA ABSOLUTA PREÇO REVENDA'] = price_dif_rev
new_df_gas['DIFEREÇA ABSOLUTA PREÇO DISTRIBUIÇÃO'] = price_dif_dist
new_df_gas['PERCENTUAL PREÇO REVENDA'] = perc_rev
new_df_gas['PERCENTUAL PREÇO DISTRIBUIÇÃO'] = perc_dist
```

```python
new_df_gas.sort_values(by='DATA INICIAL', ascending=True, inplace=True)
```

```python
new_df_gas.set_index('DATA INICIAL', inplace=True)
```
